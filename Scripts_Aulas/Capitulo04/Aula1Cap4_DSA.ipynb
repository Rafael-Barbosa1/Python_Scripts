## Lendo o banco de dados em colunas

```python
f = open('Arquivos/salarios.csv', 'r')
```

```python
dados = f.read()
```

```python
linhas = dados.split('\n')
```

```python
banco = []
```

```python
for linha in linhas:
    split_linhas = linha.split(',')
    banco.append(split_linhas)
```

```python
count = 0

for linhas in banco:
    count += 1
    
    
count
```
